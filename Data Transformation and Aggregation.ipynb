```python
#Task 2.2: Grouping and Aggregation
#Group the data by source station and count the number of trains originating from each station.
#Find the average number of trains per day for each source station.     
```

```python
train_counts_by_source = df.groupby('Source_Station_Name').size().reset_index(name='train_count')
train_counts_by_source.head()
```

```python
# Create a new column to count number of entries per day per source
df['day_count'] = 1

# Pivot to count trains per day per source
pivot = df.pivot_table(index='Source_Station_Name', columns='days', values='day_count', aggfunc='sum', fill_value=0)

# Calculate average number of trains per day
avg_trains_per_day = pivot.mean(axis=1).reset_index(name='avg_trains_per_day')
avg_trains_per_day.head()
```

```python
#Task 2.3: Data Enrichment
#Add a new column that categorizes trains based on their operating days (e.g., 'Weekday', 'Weekend')
```

```python
# Define weekday/weekend sets
weekdays = {'monday', 'tuesday', 'wednesday', 'thursday', 'friday'}
weekends = {'saturday', 'sunday'}

# Create a new column
df['operation_category'] = df['days'].str.strip().str.lower().apply(
    lambda x: 'Weekday' if x in weekdays else ('Weekend' if x in weekends else 'Other')
)

df[['Train_No', 'Train_Name', 'days', 'operation_category']].head()
```
